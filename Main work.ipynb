```python
def modular_exponentiation():

    base = int(input("Enter a base: "))
    exponent = int(input("Enter an exponent: "))
    modulus = int(input("Enter a modulus: "))

    x = 1
    power = base % modulus

    bin_exponent = bin(exponent)

    actual_bin_exponent = bin_exponent[2:]

    reverse_bin_exponent = reversed(actual_bin_exponent)
    actual_reverse_bin_exponent = "".join(reverse_bin_exponent)

    for i in range(len(actual_reverse_bin_exponent)):
        if int(actual_reverse_bin_exponent[i]) == 1:   
            x = (x * power) % modulus
        power = (power * power) % modulus
        
    print(f"\nAnswer: {base}^{exponent} mod {modulus} is {x}")

modular_exponentiation()
```

```python
import random
import string
import time
import itertools
import matplotlib.pyplot as plt


CHARACTERS = string.ascii_uppercase + string.digits 



def generate_pin():
    return "".join(random.choices(CHARACTERS, k=4))



def brute_force_crack(pin):
    start_time = time.time()
    
    for guess in ("".join(chars) for chars in itertools.product(CHARACTERS, repeat=4)):
        if guess == pin:
            time_taken = time.time() - start_time
            print(f"Time taken: {time_taken:.4f} seconds")
            return time_taken  



def main(n=500):
    crack_times = []

    for _ in range(n):
        pin = generate_pin()
        print(f"Generated PIN: {pin}")  
        time_taken = brute_force_crack(pin)
        crack_times.append(time_taken)

    return crack_times



crack_times = main()

plt.hist(crack_times, bins=5, color='green', edgecolor='black', alpha=0.7)
plt.xlabel("Time (seconds)")
plt.ylabel("Frequency")
plt.title("A Histogram Showing The Time Taken To Crack Pincodes")
plt.show()
```
